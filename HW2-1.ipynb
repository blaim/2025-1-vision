```python
from google.colab import files
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 이미지 업로드(코랩 이미지 업로드 기능 이용)
uploaded = files.upload()

filename = list(uploaded.keys())[0]

image = cv2.imread(filename, cv2.IMREAD_GRAYSCALE)

# 수업시간에 배운 power low transfortation 을 이용해 변환
def power_law_transformation(img, gamma):
    normalized_img = img / 255.0
    corrected_img = np.power(normalized_img, gamma)
    return np.uint8(corrected_img * 255)

# 감마 값 설정값이 높을수록 이미지가 선명해짐
gamma = 2.0 #ex)3.0 4.0...
corrected_image = power_law_transformation(image, gamma)

# 결과 시각화(코랩에선 cv.imshow 실행안되서 plt사용)
plt.imshow(corrected_image, cmap='gray')
plt.show()
```

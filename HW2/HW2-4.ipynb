```python
from google.colab import files
import cv2
import numpy as np
import matplotlib.pyplot as plt


#  (    )
uploaded = files.upload()

filename = list(uploaded.keys())[0]

image = cv2.imread(filename)

# Sobel      (45, -45 )
sobel_45 = np.array([
    [0,  1,  2],
    [-1, 0,  1],
    [-2, -1, 0]
])

sobel_min45 = np.array([
    [2,  1,  0],
    [1,  0, -1],
    [0, -1, -2]
])

#  (sobel   45, -45 )
edge_45 = cv2.filter2D(image, -1, sobel_45)
edge_min45 = cv2.filter2D(image, -1, sobel_min45)


#  ( 45,  -45)
plt.imshow(edge_45, cmap='gray')
plt.show()
plt.imshow(edge_min45, cmap='gray')
plt.show()

```

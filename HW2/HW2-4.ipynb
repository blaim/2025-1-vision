```python
from google.colab import files
import cv2
import numpy as np
import matplotlib.pyplot as plt


# 이미지 업로드(코랩 이미지 업로드 기능 이용)
uploaded = files.upload()

filename = list(uploaded.keys())[0]

image = cv2.imread(filename)

# Sobel 커널 변경시키기 위해 배열 설정 (45도, -45도 방향)
sobel_45 = np.array([
    [0,  1,  2],
    [-1, 0,  1],
    [-2, -1, 0]
])

sobel_min45 = np.array([
    [2,  1,  0],
    [1,  0, -1],
    [0, -1, -2]
])

# 필터 적용(sobel에 내가 만든 45도, -45도 적용)
edge_45 = cv2.filter2D(image, -1, sobel_45)
edge_min45 = cv2.filter2D(image, -1, sobel_min45)


# 결과 시각화(처음 45도, 두번째 -45)
plt.imshow(edge_45, cmap='gray')
plt.show()
plt.imshow(edge_min45, cmap='gray')
plt.show()

```

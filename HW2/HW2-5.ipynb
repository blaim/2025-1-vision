```python
from google.colab import files
import cv2
import numpy as np
import matplotlib.pyplot as plt


# 이미지 업로드(코랩 이미지 업로드 기능 이용)
uploaded = files.upload()

filename = list(uploaded.keys())[0]

image = cv2.imread(filename, cv2.IMREAD_GRAYSCALE)

#4*4로하면 지문도 뭉게져서 3*3으로 설정
se = np.uint8([
    [0,1,0],
    [1,1,1],
    [0,1,0]
])

# opening위해 이진화
t,bin_image = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)

#흰색노이즈 제거하기위해 opening사용(침식후 팽창)
eroded = cv2.erode(bin_image, se, iterations=1)
opening = cv2.dilate(eroded, se, iterations=1)

# 결과 시각화
plt.imshow(opening, cmap='gray')
plt.show()

```

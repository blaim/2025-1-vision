```python
from google.colab import files
import cv2
import numpy as np
import matplotlib.pyplot as plt


#  (    )
uploaded = files.upload()

filename = list(uploaded.keys())[0]

image = cv2.imread(filename)

#  
first_array = np.array([
    [1/9, 1/9, 1/9],
    [1/9, 1/9, 1/9],
    [1/9, 1/9, 1/9]
])

second_array = np.array([
    [0.0030, 0.0133, 0.0219, 0.0133, 0.0030],
    [0.0133, 0.0596, 0.0983, 0.0596, 0.0133],
    [0.0219, 0.0983, 0.1621, 0.0983, 0.0219],
    [0.0133, 0.0596, 0.0983, 0.0596, 0.0133],
    [0.0030, 0.0133, 0.0219, 0.0133, 0.0030]
])

third_array = np.array([
    [ 0, -1,  0],
    [-1,  4, -1],
    [ 0, -1,  0]
])

forth_array = np.array([
    [-1, -1, -1],
    [-1,  8, -1],
    [-1, -1, -1]
])

fifth_array = np.array([
    [-1, 0, 1],
    [-1, 0, 1],
    [-1, 0, 1]
])

sixth_array = np.array([
    [-1, -1, -1],
    [ 0,  0,  0],
    [ 1,  1,  1]
])

# 
gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
#    
gray16=np.int16(gray)
#   
emboss1=np.uint8(np.clip(cv2.filter2D(gray16,-1,first_array)+128,0,255))
emboss2=np.uint8(np.clip(cv2.filter2D(gray16,-1,second_array)+128,0,255))
emboss3=np.uint8(np.clip(cv2.filter2D(gray16,-1,third_array)+128,0,255))
emboss4=np.uint8(np.clip(cv2.filter2D(gray16,-1,forth_array)+128,0,255))
emboss5=np.uint8(np.clip(cv2.filter2D(gray16,-1,fifth_array)+128,0,255))
emboss6=np.uint8(np.clip(cv2.filter2D(gray16,-1,sixth_array)+128,0,255))


#  ( cv.imshow  plt)
plt.imshow(emboss1, cmap='gray')
plt.show()
plt.imshow(emboss2, cmap='gray')
plt.show()
plt.imshow(emboss3, cmap='gray')
plt.show()
plt.imshow(emboss4, cmap='gray')
plt.show()
plt.imshow(emboss5, cmap='gray')
plt.show()
plt.imshow(emboss6, cmap='gray')
plt.show()
```

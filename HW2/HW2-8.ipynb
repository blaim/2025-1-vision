```python
from google.colab import files
import cv2
import numpy as np
import matplotlib.pyplot as plt


#  (    )
uploaded = files.upload()

filename = list(uploaded.keys())[0]

image = cv2.imread(filename, cv2.IMREAD_GRAYSCALE)
######
##morphological smoothing
#####

se = np.uint8([
    [0,1,0],
    [1,1,1],
    [0,1,0]
])

#    
eroded = cv2.erode(image, se, iterations=1)
opening = cv2.dilate(eroded, se, iterations=1)

dilate = cv2.dilate(opening, se, iterations=1)
closing = cv2.erode(dilate, se, iterations=1)

####
#smoothing filter
####
#GOF  gausian filter 
smoothing = cv2.GaussianBlur(image, (5,5), 0)

#morphological smoothing 
plt.imshow(closing,cmap='gray')
plt.xlabel('MORPHOLOGICAL')
plt.show()

#smoothing filter 
plt.imshow(smoothing,cmap='gray')
plt.xlabel('SMOOTHING_FILTER')
plt.show()

###morphological smoothing     
```
